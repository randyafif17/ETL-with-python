```python
import mysql.connector
from mysql.connector import Error
import pandas as pd
import numpy as np

def extract_from_mysql(db_params, table_name):
    try:
        connection = mysql.connector.connect(**db_params)
        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql(query, connection)
        return df
    except Error as error:
        print(f"Error: {error}")
    finally:
        if connection and connection.is_connected():
            connection.close()

if __name__ == "__main__":
    db_params = {
        "host": "localhost",
        "user": "root",
        "passwd": "YOURPASSWORD",
        "database": "new"
    }
    table_name = "tables"
    data_mysql = extract_from_mysql(db_params, table_name)

    
data_mysql.to_csv('extracted_data_mysql.csv', index=False)
print(data_mysql)
```

```python
import pandas as pd
import json

def extract_from_json(file_path):
    try:
        with open(file_path, 'r') as file:
            json_data = json.load(file)
            df = pd.json_normalize(json_data)
        return df
    except Exception as e:
        print(f"Error reading JSON file: {e}")

if __name__ == "__main__":
    file_path = 'generated_data.json'
    data_json = extract_from_json(file_path)
    data_json.to_csv('extracted_data_json.csv', index=False)

print(data_json)
```

```python
import pandas as pd

def extract_from_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except Exception as e:
        print(f"Error reading CSV file: {e}")

if __name__ == "__main__":
    file_path = 'combined_data.csv'
    data_csv = extract_from_csv(file_path)
    data_csv.to_csv('extracted_data_csv.csv', index=False)
print(data_csv)
```

```python
import csv
import json
import mysql.connector
from mysql.connector import Error

def read_from_mysql(db_params, table_name):
    connection = None
    cursor = None
    try:
        # Connect to the database
        connection = mysql.connector.connect(**db_params)
        cursor = connection.cursor()
        
        # Execute the query to fetch all rows from the specified table
        cursor.execute(f"SELECT * FROM {table_name}")
        records = cursor.fetchall()
        
        # Iterate over the records and print each row
        for row in records:
            print(row)

    except mysql.connector.Error as err:
        print(f"Database error occurred: {err}")
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection and connection.is_connected():
            connection.close()
            print("MySQL connection is closed")
            
# CSV File Reading
def read_csv_file(file_path):
    try:
        with open(file_path, mode='r') as file:
            csv_reader = csv.reader(file)
            for row in csv_reader:
                print(row)
    except FileNotFoundError:
        print("The CSV file was not found.")
    except csv.Error as e:
        print(f"CSV file error: {e}")
    except Exception as e:
        print(f"An error occurred while reading the CSV file: {e}")

# JSON File Reading
def read_json_file(file_path):
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
            print(data)
    except FileNotFoundError:
        print("The JSON file was not found.")
    except json.JSONDecodeError:
        print("JSON decode error - the file may not be in proper JSON format.")
    except Exception as e:
        print(f"An error occurred while reading the JSON file: {e}")
        
if __name__ == "__main__":
    read_csv_file('combined_data.csv')
    read_json_file('generated_data.json')
    db_params = {
            "host": "localhost",
            "user": "root",
            "passwd": "YOURPASSWORD",
            "database": "new"
        }
    table_name = "data_from_sql"
    read_from_mysql(db_params, table_name)
```

```python
import pandas as pd

def transform_data(df):
    # Contoh transformasi: membersihkan spasi berlebih pada kolom string
    str_cols = df.select_dtypes(include=['object'])
    df[str_cols.columns] = str_cols.apply(lambda x: x.str.strip())
    
    # Mengubah format tanggal jika ada kolom tanggal
    date_cols = df.select_dtypes(include=['datetime64'])
    df[date_cols.columns] = date_cols.apply(pd.to_datetime, errors='coerce')
    
    # Menghitung kolom baru, misalnya 'age' dari 'birth_date'
    if 'birth_date' in df.columns:
        df['age'] = (pd.Timestamp('now') - pd.to_datetime(df['birth_date'])).dt.days // 365
    
    df = df[df['purchase_amount'] >= 0]
    # rejected_data = df[df.isnull().any(axis=1)]

    return df

# Contoh penggunaan fungsi transformasi
if __name__ == "__main__":
    # Asumsikan df_csv, df_json, df_mysql telah di-load dari masing-masing sumber
    df_csv = pd.read_csv('combined_data.csv')
    df_json = pd.read_json('generated_data.json')
    df_mysql = pd.read_csv('extracted_data_mysql.csv')  # Misalnya setelah diexport dari SQL ke CSV

    # Transform data dari masing-masing DataFrame
    df_csv_transformed = transform_data(df_csv)
    df_json_transformed = transform_data(df_json)
    df_mysql_transformed = transform_data(df_mysql)

    # Hasil transformasi bisa disimpan kembali atau digunakan untuk analisis lebih lanjut
    df_csv_transformed.to_csv('transformed_data_csv.csv', index=False)
    df_json_transformed.to_csv('transformed_data_json.csv', index=False)
    df_mysql_transformed.to_csv('transformed_data_mysql.csv', index=False)
    # df_rejected.to_csv('rejected_data.csv', index=False)

```
